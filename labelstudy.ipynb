```python
import math
import numpy as np

def decile_map(percent_total_values):
  result = [0.] * 1000
  for index, percent, total in percent_total_values:
    result[index] = math.floor(percent / 10.)
  return [(i,v) for i, v in enumerate(result)]

def compare_decile_maps(map1, map2):
  differences = np.asarray([abs(v1[1]-v2[1]) for v1, v2 in zip(map1, map2)])
  l2_diff = np.sum(differences ** 2)
  avg = np.mean(differences)
  return l2_diff, avg
    
```

```python
def plot_ciles(percent_total_values, n_ciles):
  y = [0.] * n_ciles
  bin_percent = 100. / n_ciles
  for index, percent, total in percent_total_values:
    bin_index = int(percent / bin_percent)
    y[bin_index] += 1
  x = list(range(0, 100, int(bin_percent)))
  plt.plot(x, y)
  plt.show()
```
